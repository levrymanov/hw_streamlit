```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

```python
data = pd.read_csv('temperature_data.csv', parse_dates=['timestamp'])
stats = data.groupby(['city', 'season'])['temperature'].agg(['mean', 'std']).reset_index()

data = data.merge(stats, on=['city', 'season'])

data['lower_bound'] = data['mean'] - 2 * data['std']
data['upper_bound'] = data['mean'] + 2 * data['std']

anomalies = data[(data['temperature'] < data['lower_bound']) | (data['temperature'] > data['upper_bound'])]

anomalies.to_csv('temperature_anomalies.csv', index=False)

anomalies[['city', 'timestamp', 'season', 'temperature', 'lower_bound', 'upper_bound']].head()
```

```python
import matplotlib.pyplot as plt

# Загружаем данные
data = pd.read_csv('temperature_data.csv', parse_dates=['timestamp'])

# Упорядочиваем данные
data = data.sort_values(['city', 'timestamp'])

# Добавляем скользящее среднее с окном в 365 дней (1 год)
data['temperature_ma_365'] = data.groupby('city')['temperature'].transform(lambda x: x.rolling(window=365, min_periods=30).mean())

# Визуализация трендов для одного города (например, Нью-Йорк)
plt.figure(figsize=(12, 6))
city = "New York"
subset = data[data['city'] == city]
plt.plot(subset['timestamp'], subset['temperature'], alpha=0.3, label="Температура (сырые данные)")
plt.plot(subset['timestamp'], subset['temperature_ma_365'], color='red', label="Годовое скользящее среднее")

plt.title(f"Тренд температуры в {city}")
plt.xlabel("Дата")
plt.ylabel("Температура (°C)")
plt.legend()
plt.show()
```

```python
from scipy.stats import linregress

# Преобразуем даты в числовые значения (количество дней с начала отсчёта)
data['days_since_start'] = (data['timestamp'] - data['timestamp'].min()).dt.days

# Вычисляем тренд для каждого города
trends = []
for city in data['city'].unique():
    city_data = data[data['city'] == city]

    # Выполняем линейную регрессию: temperature = slope * days + intercept
    slope, intercept, r_value, p_value, std_err = linregress(city_data['days_since_start'], city_data['temperature'])

    trends.append({"city": city, "slope": slope, "p_value": p_value, "r_squared": r_value ** 2})

# Преобразуем в DataFrame и сортируем по величине тренда
trends_df = pd.DataFrame(trends).sort_values(by="slope", ascending=False)

print(trends_df)
```

```python
import multiprocessing

# Чтение данных
data = pd.read_csv('temperature_data.csv', parse_dates=['timestamp'])

# Функция для вычисления статистики для каждого города
def calculate_seasonal_stats(group):
    return group.groupby(['season'])['temperature'].agg(['mean', 'std']).reset_index()

# Измерим время выполнения с распараллеливанием
start_time = time.time()

# Разбиваем данные по городам
grouped = data.groupby('city')

# Используем Pool для распараллеливания
with multiprocessing.Pool(processes=multiprocessing.cpu_count()) as pool:
    # Распараллеливаем вычисления
    results = pool.map(calculate_seasonal_stats, [group for _, group in grouped])

# Объединяем результаты в один DataFrame
seasonal_stats_parallel = pd.concat(results, keys=grouped.groups.keys(), names=['city']).reset_index()

end_time = time.time()

# Выводим время выполнения
print(f"Время выполнения с распараллеливанием: {end_time - start_time:.4f} секунд")

# Выводим результат
print(seasonal_stats_parallel)
```
